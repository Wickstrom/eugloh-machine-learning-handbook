## 6. Predicted vs actual

A useful diagnostic for regression is to plot `y_pred` against `y_test`. For
a perfect model all points lie on the diagonal $y = \hat{y}$. Deviations
from the diagonal tell you about the model's errors.

```python
fig, ax = plt.subplots(figsize=(6, 6))
ax.scatter(y_test, y_pred, alpha=0.3, s=10, color="#2563eb")
lims = [0, max(y_test.max(), y_pred.max())]
ax.plot(lims, lims, color="#dc2626", linestyle="--", linewidth=1.5, label="y = y_hat")
ax.set_xlabel("Actual (y_test)")
ax.set_ylabel("Predicted (y_pred)")
ax.set_title("Predicted vs actual — LinearRegression")
ax.set_xlim(lims)
ax.set_ylim(lims)
ax.set_aspect("equal")
ax.legend()
plt.show()
```

## 7. (Bonus) Ridge with feature scaling

Features are on very different scales. Standardising them and using
[Ridge regression](https://scikit-learn.org/stable/modules/linear_model.html#ridge-regression)
with a small regularisation strength often helps.

Try `Ridge(alpha=1.0)` inside a `make_pipeline(StandardScaler(), Ridge(...))`
and compare the test metrics to the baseline. Try a few values of `alpha`
(e.g. `0.01`, `1.0`, `100.0`) and see how $R^2$ changes.

```python
# YOUR CODE HERE: fit Ridge models with different alpha values and report test R^2.
for alpha in [0.01, 1.0, 100.0]:
    pipe = make_pipeline(StandardScaler(), Ridge(alpha=alpha))
    pipe.fit(X_train, y_train)
    score = pipe.score(X_test, y_test)
    print(f"alpha={alpha:>6.2f}: R^2 = {score:.4f}")
```
