## 1. Generate two clusters

Draw 60 points from each of two Gaussian distributions centred at $(-1, -1)$
and $(1.5, 1.5)$ with standard deviation $1$ on both axes.

```python
n_per_class = 60
X_pos = rng.normal(loc=[1.5, 1.5], scale=[1.0, 1.0], size=(n_per_class, 2))
X_neg = rng.normal(loc=[-1.0, -1.0], scale=[1.0, 1.0], size=(n_per_class, 2))
X = np.vstack([X_pos, X_neg])
y = np.concatenate([np.ones(n_per_class), np.zeros(n_per_class)])

# Expected: X.shape == (120, 2), y.shape == (120,), y.mean() == 0.5
```

## 2. Visualise the data

```python
fig, ax = plt.subplots(figsize=(6, 5))
ax.scatter(X_pos[:, 0], X_pos[:, 1], color="#16a34a", alpha=0.7, label="class 1")
ax.scatter(X_neg[:, 0], X_neg[:, 1], color="#dc2626", alpha=0.7, label="class 0")
ax.set_xlabel("$x_1$")
ax.set_ylabel("$x_2$")
ax.set_title("Two Gaussian clusters")
ax.legend()
ax.set_aspect("equal")
plt.show()
```

## 6. Visualise the decision boundary

We evaluate the trained model on a dense grid and colour each cell by the
predicted class. The boundary between the two coloured regions is the
**decision boundary** — a straight line, because logistic regression is a
linear classifier.

```python
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(
    np.linspace(x_min, x_max, 300),
    np.linspace(y_min, y_max, 300),
)
Z = clf.predict(np.c_[xx.ravel(), yy.ravel()]).reshape(xx.shape)

fig, ax = plt.subplots(figsize=(6, 5))
ax.contourf(xx, yy, Z, alpha=0.25, cmap="RdYlGn")
ax.scatter(X[y == 1, 0], X[y == 1, 1], color="#16a34a", alpha=0.8, label="class 1")
ax.scatter(X[y == 0, 0], X[y == 0, 1], color="#dc2626", alpha=0.8, label="class 0")
ax.set_xlabel("$x_1$")
ax.set_ylabel("$x_2$")
ax.set_title("Logistic regression decision boundary")
ax.legend()
ax.set_aspect("equal")
plt.show()
```
